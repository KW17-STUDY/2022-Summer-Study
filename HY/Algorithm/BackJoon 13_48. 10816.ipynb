```python
N = int(input())
arr1 = list(map(int, input().split()))
M = int(input())
arr2 = list(map(int, input().split()))

arr1.sort()

left = 0
right = N-1
cnt = 0 
result = []
counts = {}
for i in arr1:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1

for i in arr2:
    while left <= right:
        mid = (left + right) // 2
        if arr1[mid] == i:
            result.append(counts[i])
            break
        elif arr1[mid] > i:
            right = mid-1
        else:
            left = mid + 1
    if left>right:
        result.append(0)
    cnt = 0
    left = 0
    right = N-1
            
for i in result:
    print(i, end = " ")
```
