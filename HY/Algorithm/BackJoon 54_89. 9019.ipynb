# 첫 시도
틀린 이유: 덱에 두개의 값을 넣어서 그런거 같음
문자열과 방문 지역을 따로 관리하지 않고 하나로 관리하니 문제가 해결

```python
import sys
from collections import deque

input = sys.stdin.readline

def Left(value):
    fth = value // 1000
    etc = value % 1000
    result = etc*10 + fth
    return result

def Right(value):
    fth = value % 10
    etc = value // 10
    result = fth* 1000 + etc
    return result



T = int(input())
for _ in range(T):
    A, B = map(int, input().split())

    Map = [False for _ in range(10000)]
    queue = deque()

    queue.append((A, ""))

    while queue:
        Reg, Str = queue.popleft()
        Map[Reg] = True
        if Reg == B:
            print(Str)
            break

        
        temp = 2*Reg if 2*Reg < 10000 else 2*Reg % 10000
        if not Map[temp]:
            queue.append((temp, Str + "D"))
        temp = Reg - 1 if Reg-1 >0 else 9999
        if not Map[temp]:
            queue.append((temp, Str + "S"))
        temp = Left(Reg)
        if not Map[temp]:
            queue.append((temp, Str + "L"))
        temp = Right(Reg)
        if not Map[temp]:
            queue.append((Right(Reg), Str + "R"))
```
