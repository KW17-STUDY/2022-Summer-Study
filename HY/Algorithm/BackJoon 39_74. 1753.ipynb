```python
import heapq
import sys 
from collections import defaultdict

INF = int(1e9)

input = sys.stdin.readline

V, E = map(int, input().split())
S = int(input())

Map = defaultdict(list)
mindist = [INF for _ in range(V+1)]

visited = [False for _ in range(V+1)]

for _ in range(E):
    u, v, w = map(int, input().split())
    Map[u].append((v, w))

def dijkstra(Map, start): # 인자로 mindist 넘겨주니까 틀린거 해결
    queue = []
    mindist[start] = 0
    heapq.heappush(queue, [mindist[start], start])

    while queue:
        cur_dist, cur_dest = heapq.heappop(queue)

        if mindist[cur_dest] < cur_dist:
            continue

        for new_dest, new_dist in Map[cur_dest]:
            distance = cur_dist + new_dist
            if distance < mindist[new_dest] and not visited[new_dest]:
                mindist[new_dest] = distance
                visited[new_dest] = True
                heapq.heappush(queue, [distance, new_dest])


dijkstra(Map, S)

for i in range(1, V+1):
    if mindist[i] == INF:
        print('INF')
    else:
        print(mindist[i])




```

```python
# 정답코드 
import heapq
import sys 
from collections import defaultdict

INF = int(1e9)

input = sys.stdin.readline

V, E = map(int, input().split())
S = int(input())

Map = defaultdict(list)
mindist = [INF for _ in range(V+1)]


for _ in range(E):
    u, v, w = map(int, input().split())
    Map[u].append((v, w))

def dijkstra(Map, start, mindist):
    queue = []
    mindist[start] = 0
    heapq.heappush(queue, [mindist[start], start])

    while queue:
        cur_dist, cur_dest = heapq.heappop(queue)

        if mindist[cur_dest] < cur_dist:
            continue

        for new_dest, new_dist in Map[cur_dest]:
            distance = cur_dist + new_dist
            if distance < mindist[new_dest] :
                mindist[new_dest] = distance
                heapq.heappush(queue, (distance, new_dest))

dijkstra(Map, S, mindist)

for i in range(1, V+1):
    if mindist[i] == INF:
        print('INF')
    else:
        print(mindist[i])




```
