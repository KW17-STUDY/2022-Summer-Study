```python
import heapq
import sys
from collections import defaultdict
INF = float('inf')
T = int(input())
 
for _ in range(T):
    n, m, t = map(int, input().split())
    s, g, h = map(int, input().split())
    Map = defaultdict(list)

    for _ in range(m):
        a, b, d = map(int, input().split())
        Map[a] += [(b, d)]
        Map[b] += [(a, d)]
    candi = []

    for _ in range(t):
        candi.append(int(input()))
    candi.sort()
    dest = [[INF, False] for _ in range(n+1)] # dest[i][0]은 거리, dest[i][1] 은 g-h or h-g를 지났는지 확인

    queue = []

    dest[s][0] = 0
    heapq.heappush(queue, (dest[s][0], s))

    while queue:
        cur_dist, cur_dest = heapq.heappop(queue)

        if cur_dist > dest[cur_dest][0]:
            continue
        
        for temp_dest, temp_dist in Map[cur_dest]:
            new_dist = cur_dist + temp_dist
            new_check = dest[cur_dest][1]
            if new_dist <= dest[temp_dest][0]: # 최단거리가 2개이상일 수 도 있기에 <=로 설정
                dest[temp_dest][0] = new_dist
                heapq.heappush(queue, (new_dist, temp_dest))
                if (cur_dest, temp_dest) == (g, h) or \
                    (cur_dest, temp_dest) == (h, g) : # g, h를 지났거나 이전에 지나왔으면 True
                    new_check = True
                dest[temp_dest][1] = new_check

    result = []
    for i in candi:
        if dest[i][1] == True:
            result.append(i)

    print(" ".join(str(i) for i in result))

        
```
