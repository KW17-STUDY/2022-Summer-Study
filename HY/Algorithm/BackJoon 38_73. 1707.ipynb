```python
from collections import deque
from collections import defaultdict
import sys

input = sys.stdin.readline

K = int(input())

def bfs(Map, visited, start):
    queue = deque()
    queue.append(start)
    visited[start] = [True, 0]
    
    while queue:
        node = queue.popleft()
        
        col = visited[node][1]
        for i in Map[node]:
            check, new_col = visited[i]
            if check == False:
                queue.append(i)
                visited[i] = [True, (col+1)%2]
            elif check == True:
                if col == new_col:
                    print("NO")
                    return False
            
    return True

for _ in range(K):
    V, E = map(int, input().split())
    Map = defaultdict(list)
    visited = [[False, 0] for _ in range(V+1)]
    count = 0
    for _ in range(E):
        u,v = map(int, input().split())
        Map[u] += [v]
        Map[v] += [u]
    for i in range(1, V+1):
        YoN = 0
        if visited[i][0] == False:
            YoN = bfs(Map, visited, i)
            if YoN == False:
                count = 1
                break
        
    if count == 0:
        print("YES")
        
    
```

```python
#bfs 부분이 무언가 틀림

from collections import deque
from collections import defaultdict
import sys
input = sys.stdin.readline

K = int(input())

def bfs(Map, visited, start):
    queue = deque()
    queue.append(start)
    if visited[start] == 0:
        visited[start] = 1
    
    while queue:
        node = queue.popleft()
        
        
        for i in Map[node]:
            if visited[i] == 0:
                queue.append(i)
                if visited[node] == 1:
                    visited[i] = 2
                else:
                    visited[i] = 1
            elif visited[i] == 1:
                if visited[node] == 1:
                    print("NO")
                    return False
            elif visited[i] == 2:
                if visited[node] == 2:
                    print("NO")
                    return False
            
    return True

for _ in range(K):
    V, E = map(int, input().split())
    Map = defaultdict(list)
    visited = [0 for _ in range(V+1)]
    count = 0
    for _ in range(E):
        a,b = map(int, input().split())
        Map[a] += [b]
        Map[b] += [a]
    for k in range(1, V+1):
        
        if not bfs(Map, visited, k) :
            count = 1
            break
        
    if count == 0:
        print("YES")
        
    
```
